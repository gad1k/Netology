```python
documents = [
    {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
    {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
    {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
]

directories = {
    "1": ["2207 876234", "11-2"],
    "2": ["10006"],
    "3": []
}


def show_name():
    number = input("Введите номер документа: ")

    for doc in documents:
        if doc["number"] == number:
            print("Владелец документа:", doc["name"])
            return

    print("Документ не найден в базе")


def show_store():
    number = input("Введите номер документа: ")

    for store, docs in directories.items():
        if number in docs:
            print("Документ хранится на полке:", store)
            return

    print("Документ не найден в базе")


def show_full():
    for doc in documents:
        cur_store = ""
        for store, docs in directories.items():
            if doc["number"] in docs:
                cur_store = store
        print(f'№: {doc["number"]}, тип: {doc["type"]}, владелец: {doc["name"]}, полка хранения {cur_store}')


def add_store():
    store = input("Введите номер полки: ")

    if not store.isnumeric() or int(store) <= 0:
        print("Введен неверный номер полки")
        return

    if store in directories:
        print("Такая полка уже существует.", end=" ")
    else:
        directories[store] = list()
        print("Полка добавлена.", end=" ")

    print("Текущий перечень полок:", ", ".join(sorted(directories.keys(), key=int)))


def delete_store():
    store = input("Введите номер полки: ")

    if store in directories.keys():
        if len(directories[store]) > 0:
            print("На полке есть документы, удалите их перед удалением полки.", end=" ")
        else:
            print("Полка удалена.", end=" ")
            directories.pop(store)
    else:
        print("Такой полки не существует.", end=" ")

    print("Текущий перечень полок:", ", ".join(sorted(directories.keys(), key=int)))


def add_doc():
    number = input("Введите номер документа: ")
    doc_type = input("Введите тип документа: ")
    name = input("Введите владельца документа: ")
    store = input("Введите полку для хранения: ")

    if store in directories.keys():
        documents.append({"type": doc_type, "number": number, "name": name})
        directories[store].append(number)
        print("Документ добавлен.", end=" ")
    else:
        print("Такой полки не существует. Добавьте полку командой as.")

    print("Текущий список документов:")
    show_full()


def delete_doc():
    number = input("Введите номер документа: ")
    document = [doc for doc in documents if doc["number"] == number]

    if len(document) == 1:
        documents.remove(document[0])
        print("Документ удален.")
    else:
        print("Документ не найден в базе.")

    for docs in directories.values():
        if number in docs:
            docs.remove(number)

    print("Текущий список документов:")
    show_full()


def move_doc():
    number = input("Введите номер документа: ")
    store = input("Введите номер полки: ")

    is_exists = False
    for doc in documents:
        if doc["number"] == number:
            is_exists = True
    if not is_exists:
        print("Документ не найден в базе.")
        print("Текущий список документов:")
        show_full()
        return

    if store not in directories.keys():
        print("Такой полки не существует.", end=" ")
        print("Текущий перечень полок:", ", ".join(sorted(directories.keys(), key=int)))
        return

    for docs in directories.values():
        if number in docs:
            docs.remove(number)

    directories[store].append(number)

    print("Документ перемещен.")
    print("Текущий список документов:")
    show_full()


while True:
    cmd = input("Введите команду: ").lower()

    if cmd == "q":
        break
    elif cmd == "p":
        show_name()
    elif cmd == "s":
        show_store()
    elif cmd == "l":
        show_full()
    elif cmd == "as":
        add_store()
    elif cmd == "ds":
        delete_store()
    elif cmd == "ad":
        add_doc()
    elif cmd == "d":
        delete_doc()
    elif cmd == "m":
        move_doc()
```
